```python
def missing_data(cloud):
    
    return cloud.isnull().sum().sort_values(ascending=False).to_frame(name='Total').assign(
        
        Percentage=lambda x: (x / len(cloud) * 100)
    ).style.background_gradient(cmap='Greens')

missing_data(cloud)
```

```python
q1 = np.percentile(cloud, 25)
q3 = np.percentile(cloud, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = (cloud < lower_bound) | (cloud > upper_bound)

outliers.describe()
```

```python
from scipy.stats import zscore

z_scores = zscore(cloud)
outliers = (abs(z_scores) > 10)  # You can adjust the threshold as needed
gemini = cloud[~outliers]

gemini
```

```python
def missing_data(gemini):
    
    return gemini.isnull().sum().sort_values(ascending=False).to_frame(name='Total').assign(
        
        Percentage=lambda x: (x / len(gemini) * 100)
    ).style.background_gradient(cmap='Greens')

missing_data(gemini)
```

```python
x = cloud.iloc[:,cloud.columns != 'stress_level']


x
```

```python
y = cloud.iloc[:,-1]

y
```

```python
cloud.columns
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y, test_size = 0.30, random_state = 42)

train_x
```

```python
from sklearn.linear_model import LinearRegression

terrain= LinearRegression()

terrain.fit(train_x,train_y)

zephyr7b = terrain.predict(test_x)

zephyr7b
```

```python
print(terrain.predict([[10,23,1,13,2,1,2,2,3,3,2,2,3,2,3,2,3,3,3,3]]))
```

```python
print('Slope is :')

list(zip(x,terrain.coef_))
```

```python
print('Constant is:' ,terrain.intercept_)
```

```python
dark = pd.DataFrame({'Actual_value':test_y,'Predicted_value': zephyr7b,'Difference': test_y- zephyr7b})

dark
```

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score

mape = mean_absolute_percentage_error(test_y, zephyr7b)

print('MAPE: {:.2%}'.format(mape))
```

```python
print('R2 SCORE IS :', r2_score(test_y,zephyr7b))
```
